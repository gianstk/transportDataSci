```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import numpy as np
```

```python
# open file
file_name = "ALL_OD_2020-05-22.csv"
df  = pd.read_csv(file_name, nrows=1000)
df
```

```python
# grouping file

data = []

for index, row in df.iterrows():
    timeString = row["PICKUP_DATETIME"][11:]
    hr = timeString[0:2]
    data.append(int(hr))
        
plt.hist(data, bins=np.arange(0, 24), color=(0.35,0.8,0.5,1), ec='k')
plt.xticks(range(25))
plt.title('When the sun goes down to gian')
plt.xlabel('time')
plt.ylabel('frequency')
plt.show()
```
