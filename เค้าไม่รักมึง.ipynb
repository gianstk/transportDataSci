```python
# เปิดไฟล์

import pandas as pd
import json

file_name="chaiya.csv"

# specify n rows
df = pd.read_csv(file_name, nrows = 1000)

# all data
# df = pd.read_csv(file_name)

df
```

```python
# algorithm

finalArray = []

car_running = False

tripNumber = 0
cardDriver = ""
pickupDateTime = 0
pickupLatitude = 0
pickupLongitude = 0

dropoffDateTime = 0
dropoffLatitude = 0
dropoffLongtitude = 0
meterFare = 0


for index, row in df.iterrows():
    
    # start meter
    if (not car_running and row["METER"] == "ON"):
        car_running = True
        # collect pickup data
        cardDriver = row["CARD_DRIVER"]
        pickupLatitude = row["LATITUDE"]
        pickupLongitude = row["LONGITUDE"]
        pickupDateTime = row["TIMESTAMP"]
        vehicleId = row["VEHICLE_ID"]
    

    
    if (car_running and row["METER"] == "OFF"):
        car_running = False
        # collect dropoff data
        item = {
            "TRIP_NUMBER": tripNumber,
            "VEHICAL_ID": vehicleId,
            "CARD_DRIVER": cardDriver,
            "PICKUP_DATETIME": pickupDateTime,
            "PICKUP_LATITUDE": pickupLatitude,
            "PICKUP_LONGTITUDE": pickupLongtitude,
            "DROPOFF_LATITUDE": dropoffLatitude,
            "DROPOFF_LONGTITUDE": dropoffLongtitude,
            "METER_FARE": meterFare
        }
        finalArray.append(item)
        tripNumber += 1
        
    meterFare = row["METER_FARE"]
    dropoffDateTime = row["TIMESTAMP"]
    dropoffLatitude = row["LATITUDE"]
    dropoffLongtitude = row["LONGITUDE"]
        
print(finalArray)
```

```python
# เขียนไฟล์

import csv

with open('ควยชัยตอมทิ้งแล้วทิ้งอีก.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["TRIP_NUMBER", "VEHICAL_ID", "CARD_DRIVER", "PICKUP_DATETIME", "PICKUP_LATITUDE", "PICKUP_LONGTITUDE", "DROPOFF_LATITUDE", "DROPOFF_LONGTITUDE", "METER_FARE"])
    
    for i in finalArray:
        writer.writerow([
            i["TRIP_NUMBER"],
            i["VEHICAL_ID"],
            i["CARD_DRIVER"],
            i["PICKUP_DATETIME"],
            i["PICKUP_LATITUDE"],
            i["PICKUP_LONGTITUDE"],
            i["DROPOFF_LATITUDE"],
            i["DROPOFF_LONGTITUDE"],
            i["METER_FARE"],
        ])
```
